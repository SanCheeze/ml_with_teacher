## Исследование задачи

Создадим константу, опредлеляющую все дальнейшие псевдослучайные процессы:

```python
RANDOM_STATE = 1234567
```

### Распределение данных на выборки

Разобьём данные на три выборки:
- тренировочную
- валидационную
- тестовую

Начнём с тестовой. Выделим 20% данных и сохраним их в `df_test`:

```python
df, df_test = train_test_split(df, test_size=.2, random_state=RANDOM_STATE)
```

Теперь можно разделить выборку на 2 части: `df_train` и `df_valid`:

```python
df_train, df_valid = train_test_split(df, test_size=.25, random_state=RANDOM_STATE)
```

Проверим, верно ли распределились данные:

```python
print('df.shape = \t\t', len(df))
print('df_test.shape = \t', len(df_test))
print('df_train.shape = \t', len(df_train))
print('df_valid.shape = \t', len(df_valid))
```

Перемешаем данные и разобьём выборки на ключевые и целевые признаки:

```python
df_train = shuffle(df_train, random_state=RANDOM_STATE)
df_valid = shuffle(df_valid, random_state=RANDOM_STATE)
df_test  = shuffle(df_test, random_state=RANDOM_STATE)
```

```python
# train
features_train = df_train.drop(['exited'], axis=1)
target_train = df_train['exited']

# valid
features_valid = df_valid.drop(['exited'], axis=1)
target_valid = df_valid['exited']

# test
features_test = df_test.drop(['exited'], axis=1)
target_test = df_test['exited']
```

```python
features_train.head()
```

```python
numeric_columns = ['credit_score', 'age', 'balance', 'estimated_salary', 'tenure', 'germany', 'spain']

scaler = StandardScaler().fit(features_train[numeric_columns])

features_valid[numeric_columns] = (scaler.transform(features_valid[numeric_columns]))
features_test[numeric_columns] = (scaler.transform(features_test[numeric_columns]))
features_train[numeric_columns] = (scaler.transform(features_train[numeric_columns]))
```

```python
features_train.head()
```

**Вывод:**

Порядок. Перейдём к моделям.

### Поиск лучшей модели

В качестве классификатора будем использовать `RandomForestClassifier`.

Нужно определить лучшие для модели гиперпараметры (глубину и количество деревьев).

Глубина (кол-во деревьев возьмём 20):

```python
%%time

best_depth = model_forest = None
best_accuracy = 0

for depth in range(1, 16, 1):
    model = RandomForestClassifier(n_estimators=20, max_depth=depth, random_state=RANDOM_STATE)
    model.fit(features_train, target_train)

    current_accuracy = model.score(features_valid, target_valid)

    if current_accuracy > best_accuracy:
        best_depth = depth
        best_accuracy = current_accuracy
        model_forest = model

print('Лучшая модель:\t', model_forest)
print('Глубина:\t\t', best_depth)
print('Точность:\t', best_accuracy)
print('\n----------------------------------------------------------------------------------------\n')
```

Теперь количество деревьев:

```python
%%time

best_est_amount = None
best_accuracy = 0

for est in range(1, 60):
    model = RandomForestClassifier(n_estimators=est, max_depth=best_depth, random_state=RANDOM_STATE)
    model.fit(features_train, target_train)

    current_accuracy = model.score(features_valid, target_valid)

    if current_accuracy > best_accuracy:
        best_est_amount = est
        best_accuracy = current_accuracy
        model_forest = model

print('Лучшая модель:\t', model_forest)
print('Глубина:\t\t', best_depth)
print('Деревья:\t\t', best_est_amount)
print()
print('Точность на тренировочной выборке:', round(model_forest.score(features_train, target_train), 3))
print('Точность на валидационной выборке:', round(model_forest.score(features_valid, target_valid), 3))
print('\n------------------------------------------------------------------------------------------\n')
```

    Модель в среднем обучается за 18 секунд, при этом имея точность 87% на валидационной выборке!

Прогоним через модель и тестовые, и валидационные данные сразу:

```python
print(f'Точность: {round(model_forest.score(df.drop("exited", axis=1), df["exited"]), 3)*100}%')
```

### Баланс разброса классов

#### Баланс весов классов

Теперь в модель передадим параметр `class_weight='balanced'`:

Нужно определить лучшие для модели гиперпараметры (глубину и количество деревьев).

```python
%%time

best_depth_balanced = model_forest_balanced = None
best_f1_balanced = 0

for depth in range(1, 21, 1):
    model = RandomForestClassifier(n_estimators=20, max_depth=depth, random_state=RANDOM_STATE, class_weight='balanced')
    model.fit(features_train, target_train)

    current_f1 = model.score(features_valid, target_valid)

    if current_f1 > best_f1_balanced:
        best_depth_balanced = depth
        best_f1_balanced = current_f1
        model_forest_balanced = model

print('Лучшая модель:\t', model_forest_balanced)
print('Глубина:\t\t', best_depth_balanced)
print('F1:\t', best_f1_balanced)
print('\n----------------------------------------------------------------------------------------\n')
```

Теперь количество деревьев:

```python
%%time

best_est_amount = None
best_f1_balanced = 0

for est in range(1, 60):
    model = RandomForestClassifier(n_estimators=est, max_depth=best_depth_balanced, class_weight='balanced', random_state=RANDOM_STATE)
    model.fit(features_train, target_train)

    current_f1 = f1_score(target_valid, model.predict(features_valid))

    if current_f1 > best_f1_balanced:
        best_est_amount = est
        best_f1_balanced = current_f1
        model_forest_balanced = model

print('Лучшая модель:\t', model_forest_balanced)
print('Глубина:\t\t', best_depth_balanced)
print('F1:\t', best_f1_balanced)
print('Деревья:\t\t', best_est_amount)
print()
print('Точность на тренировочной выборке:', round(model_forest_balanced.score(features_train, target_train), 3))
print('Точность на валидационной выборке:', round(model_forest_balanced.score(features_valid, target_valid), 3))
print('\n------------------------------------------------------------------------------------------\n')
```

Recall, Precision, F1

```python
recall_balanced_ = round(recall_score(target_valid, model_forest_balanced.predict(features_valid)), 3)
print('Модель верно определяет такую долю положительных объектов: '
      f'{recall_balanced_}.')

precision_balanced_ = round(precision_score(target_valid, model_forest_balanced.predict(features_valid)), 3)
print('Доля положительных объектов, которые модель обозначила положительными: '
      f'{precision_balanced_}')
```

```python
f1_balanced = round(f1_score(target_valid, model_forest_balanced.predict(features_valid)), 2)

print(f'F1-мера новой модели (баланс классов): {f1_balanced}')
```

#### Уменьшение выборки

Для реализации этого напишем функцию:

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features = pd.concat([features_zeros.sample(frac=fraction, random_state=RANDOM_STATE)] + [features_ones])
    target = pd.concat([target_zeros.sample(frac=fraction, random_state=RANDOM_STATE)] + [target_ones])
    
    features, target = shuffle(features, target, random_state=RANDOM_STATE)
    
    return features, target
```

```python
features_downsampled, target_downsampled = downsample(features_train, target_train, 0.1)

pd.DataFrame(data=[[features_train.shape[0], target_train.shape[0]], [features_downsampled.shape[0], target_downsampled.shape[0]]], columns=['features', 'target'], index=['default', 'downsampled'])
```

#### Увеличение выборки

Для реализации этого также напишем функцию:

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=RANDOM_STATE)
    
    return features_upsampled, target_upsampled
```
